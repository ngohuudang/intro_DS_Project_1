-scrollPage là hàm dùng để auto lướt trang web xuống để chúng ta có thể load dữ liệu cần lấy.

```python
def scrollPage(scrollTime,driver):
    while scrollTime!=0:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        scrollTime-=1
        time.sleep(sleep_time)
```

-use_Selenium là hàm chính dùng để mở web và thực hiện tất cả những thao tác mà chúng ta cần để có thể lấy được dữ liệu từ web.

```python
def use_Selenium(whatlist,timeScroll,genre,des):
    Options=webdriver.ChromeOptions()
    prefs = {"profile.managed_default_content_settings.images": 2}
    Options.add_experimental_option("prefs", prefs)
    Options.add_argument('--app=https://soundcloud.com/')
    driver = webdriver.Chrome(options=Options)
    driver.get(f'https://soundcloud.com/search?q={genre}')
    time.sleep(sleep_time)
    x=des*400
    y=10
    driver.set_window_rect(x,y,400,600)
    accept=driver.find_element_by_id('onetrust-accept-btn-handler')
    accept.click()
    time.sleep(sleep_time)    
    if whatlist in "playlists":
        driver.find_element_by_xpath(
            "/html/body/div[1]/div[2]/div[2]/div/div/div[2]/div/div/div/div/ul/li[5]/a"
        ).click()
        time.sleep(sleep_time)
        scrollPage(timeScroll,driver)
        get_Playlists(driver)
    elif whatlist in "tracks":
        driver.find_element_by_xpath(
            "/html/body/div[1]/div[2]/div[2]/div/div/div[2]/div/div/div/div/ul/li[2]/a"
        ).click()
        time.sleep(sleep_time)
        scrollPage(timeScroll,driver)
        get_Tracks(driver)
    elif whatlist in "users":
        driver.find_element_by_xpath(
            "/html/body/div[1]/div[2]/div[2]/div/div/div[2]/div/div/div/div/ul/li[3]/a"
        ).click()
        time.sleep(sleep_time)
        scrollPage(timeScroll,driver)
        get_Users(driver)
    driver.close()
```

-PlaylistThread là lớp được tạo ra để việc xử lý dữ liệu playlists được diễn ra một cách nhanh hơn bằng cách áp dụng đa luồng trong python.

```python
class PlaylistThread(threading.Thread):

    def __init__(self, pool_sema,playlist):
        threading.Thread.__init__(self)
        self.pool_sema = pool_sema
        self.playlist=playlist
    def run(self):
        self.pool_sema.acquire()
        try:
            url=self.playlist.find_element_by_class_name('sound__coverArt').get_attribute('href')
            userPL=self.playlist.find_element_by_class_name('soundTitle__usernameText')
            titlePL=self.playlist.find_element_by_class_name('soundTitle__title')
            uploadTime=self.playlist.find_element_by_class_name('sc-visuallyhidden')
            like=self.playlist.find_element_by_class_name('sc-button-like')
            repost=self.playlist.find_element_by_class_name('sc-button-repost')
            track_list =self.playlist.find_element_by_class_name('sound__trackList')
            tracks=track_list.find_elements_by_tag_name('li')
            TRACKS=[]
            trackcount=''
            try:
                track_count=self.playlist.find_element_by_class_name('compactTrackList__moreLink')
            except:
                trackcount=len(tracks)
            for track in tracks:
                trackId=track.find_element_by_class_name('compactTrackListItem__number')
                trackUser=track.find_element_by_class_name('compactTrackListItem__user')
                trackTitle=track.find_element_by_class_name('compactTrackListItem__trackTitle')
                TRACKS.append(trackTitle.text)
            result={}
            result['url']=url
            result['user']=userPL.text
            result['title']=titlePL.text
            result['upload_time']=uploadTime.text
            if like.text=='Like':
                result['num_likes']='0'
            else:
                result['num_likes']=like.text
            if repost.text=='Repost':
                result['num_reposts']='0'
            else:
                result['num_reposts']=repost.text
            if trackcount!='':
                result['track_count']=trackcount
            else:
                result['track_count']=track_count.text[5:-7]
            result['some_tracks']=TRACKS
            resultsPlaylist.append(result)
        except:
            ''''''
```

-get_Playlists là hàm dùng để lấy dữ liệu playlists từ web.

-write_File là hàm dùng để ghi dữ liệu playlists ra file csv.

```python
def get_Playlists(driver):
    pool_sema = threading.Semaphore(value=max_connections)
    html_list =driver.find_element_by_class_name('search__listWrapper')
    playlists=html_list.find_elements_by_class_name('searchList__item')
    for playlist in playlists:
        thread=PlaylistThread(pool_sema,playlist)
        thread.start()
        time.sleep(1)

def write_File(resultsPlaylist):
    with open('Crawl_data/playlist.csv','w+',encoding='utf-8') as f:
            w=csv.DictWriter(f,[
                'url',
                'user',
                'title',
                'upload_time',
                'num_likes',
                'num_reposts',
                'track_count',
                'some_tracks'
            ],delimiter=',')
            w.writeheader()
            for result in resultsPlaylist:
                w.writerow(result)
    f.close()
```

-TrackThread là lớp được tạo ra để việc xử lý dữ liệu tracks được diễn ra một cách nhanh hơn bằng cách áp dụng đa luồng trong python.

```python
class TrackThread(threading.Thread):

    def __init__(self, pool_sema,track):
        threading.Thread.__init__(self)
        self.pool_sema = pool_sema
        self.track=track
    def run(self):
        self.pool_sema.acquire()
        try:
            url=self.track.find_element_by_class_name(
                'sound__coverArt'
            ).get_attribute('href')
            userPL=self.track.find_element_by_class_name('soundTitle__usernameText')
            titlePL=self.track.find_element_by_class_name('soundTitle__title')
            uploadTime=self.track.find_element_by_class_name('sc-visuallyhidden')
            like=self.track.find_element_by_class_name('sc-button-like')
            repost=self.track.find_element_by_class_name('sc-button-repost')
            num_plays=self.track.find_element_by_class_name(
                'sc-ministats-plays'
            ).find_element_by_class_name('sc-visuallyhidden')
            num_cmts=self.track.find_element_by_class_name('sc-ministats-comments')
            result={}
            result['url']=url
            result['user']=userPL.text
            result['title']=titlePL.text
            result['upload_time']=uploadTime.text
            if like.text=='Like':
                result['num_likes']='0'
            else:
                result['num_likes']=like.text
            if repost.text=='Repost':
                result['num_reposts']='0'
            else:
                result['num_reposts']=repost.text
            result['num_plays']=num_plays.text[:-6]
            result['num_comments']=num_cmts.text[18:]
            resultsTrack.append(result)
        except:
            ''''''
```

-get_Tracks là hàm dùng để lấy dữ liệu tracks từ web.

-write_fileTracks là hàm dùng để ghi dữ liệu tracks ra file csv.

```python
def get_Tracks(driver):
    pool_sema2 = threading.Semaphore(value=max_connections)
    html_list_track =driver.find_element_by_class_name('search__listWrapper')
    tracks=html_list_track.find_elements_by_class_name('searchList__item')
    for track in tracks:
        thread=TrackThread(pool_sema2,track)
        thread.start()
        time.sleep(1)

def write_fileTracks(resultsTrack):
    with open('Crawl_data/track.csv','w+',encoding='utf-8') as f:
            w=csv.DictWriter(f,[
                'url',
                'user',
                'title',
                'upload_time',
                'num_likes',
                'num_reposts',
                'num_plays',
                'num_comments'
            ],delimiter=',')
            w.writeheader()
            for result in resultsTrack:
                w.writerow(result)
    f.close()
```

-UserThread là lớp được tạo ra để việc xử lý dữ liệu users được diễn ra một cách nhanh hơn bằng cách áp dụng đa luồng trong python.

```python
class UserThread(threading.Thread):

    def __init__(self, pool_sema,user):
        threading.Thread.__init__(self)
        self.pool_sema = pool_sema
        self.user=user
    def run(self):
        self.pool_sema.acquire()
        try:
            url=self.user.find_element_by_class_name(
                'userItem__coverArt'
            ).get_attribute('href')
            numtrack=''
            userInfo=self.user.find_element_by_class_name('userItem__title')
            detail=self.user.find_element_by_class_name('userItem__details')
            details=detail.text
            if details=='':
                details='None'
            number=self.user.find_element_by_class_name(
                'sc-ministats-group'
            ).find_elements_by_tag_name('li')
            if len(number)==1:
                num_followers=self.user.find_element_by_class_name(
                    'sc-ministats-followers'
                ).find_element_by_class_name('sc-visuallyhidden')
                numtrack='0'
            else:
                num_followers=self.user.find_element_by_class_name(
                    'sc-ministats-followers'
                ).find_element_by_class_name('sc-visuallyhidden')
                num_tracks=self.user.find_element_by_class_name(
                    'sc-ministats-sounds'
                ).find_element_by_class_name('sc-visuallyhidden')
            result={}
            result['url']=url
            result['user']=userInfo.text
            result['details']=details
            result['num_followers']=num_followers.text[:-10]
            if numtrack=='0':
                result['num_tracks']=numtrack
            else:
                result['num_tracks']=num_tracks.text[:-7]
            resultsUser.append(result)
        except:
            ''''''
```

#API
Hàm collect_repositories: lấy repositories trong 1 lần

Nếu không lấy được sẽ thực hiện sleep() rồi lấy lại

hàm trả về danh sách các collection chứa danh sách các playlist,user,track

```python
def collect_repositories(genre,client_id,limit_count,offset_count):
    repositories = []
    time.sleep(sleep_time)
    url_soundcloud=f'https://api-v2.soundcloud.com/search/playlists?q={genre}&client_id={client_id}&limit={limit_count}&offset={offset_count}'
    r= requests.get(url_soundcloud)
    while not r.ok:
        time.sleep(sleep_time)
        r= requests.get(url_soundcloud)   
    data = json.loads(r.text)
    for i in data['collection']:
        if i not in repositories:
            repositories.append(i)
    return repositories 
```

Hàm collect_all_repositories lấy repositories cho đến khi đủ số lượng yêu cầu

```python
def collect_all_repositories(genre,client_id,limit_count,min_repositories):
    all_repositories = []
    starting_offset=0
    while len(all_repositories)<min_repositories:
        r=collect_repositories(genre,client_id,limit_count,starting_offset)
        all_repositories.extend(r)
        starting_offset+=limit_count
    return all_repositories
```

```python
def collect_playlists_to_csv(repositories):
    playlist_id=[]
    playlist_title=[]
    playlist_user_id=[]
    playlist_description=[]
    playlist_url=[]
    playlist_genre=[]
    playlist_username=[]
    playlist_track=[]
    playlist_track_str=''
    playlist_track_count=[]
    playlist_likes_count=[]
    playlist_reposts_count=[]
    for repo in repositories:
        playlist_id.append(repo['id'])
        playlist_title.append(repo['title'])
        playlist_user_id.append(repo['user_id'])
        playlist_description.append(repo['description'])
        playlist_url.append(repo['permalink_url'])
        playlist_genre.append(repo['genre'])
        playlist_username.append(repo['user']['username'])
        head_of_tracks=20 if len(repo['tracks'])>20 else len(repo['tracks'])-1
        for i in range(head_of_tracks):
            playlist_track_str+=str(repo['tracks'][i]['id'])+' '
        playlist_track_count.append(repo['track_count'])
        playlist_likes_count.append(repo['likes_count'])
        playlist_reposts_count.append(repo['reposts_count'])
        playlist_track.append(playlist_track_str)
        playlist_track_str=''
    playlist_frame={
        'id': playlist_id,
        'title': playlist_title,
        'genre': playlist_genre,
        'description': playlist_description,
        'url': playlist_url,
        'user_id': playlist_user_id,
        'username': playlist_username,
        'trackIds':playlist_track,
        'track_count': playlist_track_count,
        'likes_count':playlist_likes_count,
        'reposts_count': playlist_reposts_count
    }
    pl=pd.DataFrame(playlist_frame)
    pl.to_csv('Api_data/playlist.csv', index=False)
```

```python
def collect_users_to_csv(repositories):
    user_id=[]
    user_city=[]
    user_country=[]
    user_description=[]
    user_followers_count=[]
    user_followings_count=[]
    user_likes_count=[]
    user_playlist_likes_count=[]
    user_url=[]
    user_playlist_count=[]
    user_track_count=[]
    user_username=[]
    for repo in repositories:
        user=repo['user']
        user_id.append(user['id'])
        user_username.append(user['username'])
        user_city.append(user['city'])
        user_country.append(user['country_code'])
        user_description.append(user['description'])
        user_followers_count.append(user['followers_count'])
        user_followings_count.append(user['followings_count'])
        user_likes_count.append(user['likes_count'])
        user_playlist_likes_count.append(user['playlist_likes_count'])
        user_url.append(user['permalink_url'])
        user_playlist_count.append(user['playlist_count'])
        user_track_count.append(user['track_count'])
    user_frame={
        'id': user_id,
        'username': user_username,
        'city': user_city,
        'country' :user_country,
        'description': user_description,
        'followers_count': user_followers_count,
        'followings_count': user_followings_count,
        'likes_count': user_likes_count,
        'playlist_likes_count': user_playlist_likes_count,
        'url': user_url,
        'playlist_count':user_playlist_count,
        'track_count':user_track_count
    }
    us=pd.DataFrame(user_frame)
    us.to_csv('Api_data/user.csv', index=False)
```

playlist: id của playlist chứa track

user_id: id của user sở hữu track

description: phần mô tả của track

comment_count: số lượt bình luận trong bài track

downloadable: có thể tải track về hay không

download_count: số lượng download track

genre: thể loại của track

likes_count: số lượt like của track

url: địa chỉ web của track

playback_count: số lượt nghe của track

reposts_count: số lượt đăng lại từ user khác của track

```python
def collect_tracks_to_csv(repositories):
    track_playlist=[]
    track_id=[]
    track_title=[]
    track_user_id=[]
    track_artist=[]
    track_description=[]
    track_comment_count=[]
    track_downloadable=[]
    track_download_count=[]
    track_genre=[]
    track_likes_count=[]
    track_url=[]
    track_playback_count=[]
    track_reposts_count=[]
    for repo in repositories:
        head_of_tracks=10 if len(repo['tracks'])>10 else len(repo['tracks'])-1
        for i in range(head_of_tracks):
            if len(repo['tracks'][i])<5:#loại những tracks không có đủ thông tin
                continue
            track_playlist.append(repo['id'])
            track=repo['tracks'][i]
            track_id.append(track['id'])
            track_title.append(track['title'])
            track_user_id.append(track['user_id'])
            if track['publisher_metadata']:
                if 'artist' in track['publisher_metadata']:
                    track_artist.append(track['publisher_metadata']['artist'])
                else: track_artist.append('')
            else: track_artist.append('')
            track_description.append(track['description'])
            track_comment_count.append(track['comment_count'])
            track_downloadable.append(track['downloadable'])
            track_download_count.append(track['download_count'])
            track_genre.append(track['genre'])
            track_likes_count.append(track['likes_count'])
            track_url.append(track['permalink_url'])
            track_playback_count.append(track['playback_count'])
            track_reposts_count.append(track['reposts_count'])
    track_frame={
        'playlist': track_playlist,
        'id': track_id,
        'title': track_title,
        'user_id': track_user_id,
        'artist': track_artist,
        'description': track_description,
        'comment_count': track_comment_count,
        'downloadable': track_downloadable,
        'download_count': track_download_count,
        'genre': track_genre, 
        'likes_count': track_likes_count, 
        'url': track_url, 
        'playback_count': track_playback_count, 
        'reposts_count': track_reposts_count
    }
    tr=pd.DataFrame(track_frame)
    tr.to_csv('Api_data/track.csv', index=False)
```
